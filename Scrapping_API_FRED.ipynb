## **Seleccionamos solo lo necesario**

```python
observations = json_response["observations"]
observations
```

# **Se crea la serie necesaria con los datos obtenidos**

```python
Fed_Funds = []
for elemento in observations:
    value = elemento.get("value")
    if value != ".":
        Fed_Funds.append({"Fecha": elemento.get("date"), "Tasa": float(value)})
```

## Check

```python
Fed_Funds
```

# **Se crea el dataframe con lo obtenido**

```python
dataframe = pd.DataFrame(Fed_Funds)
```
